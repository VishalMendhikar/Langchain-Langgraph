### Text Files(.txt)

```python
import os
os.makedirs("data/text_files", exist_ok=True)
```

```python
sample_text = {
    "data/text_files/Python.txt": "Python is a high-level programming language. It is widely used for web development, data analysis, artificial intelligence, and scientific computing.",
    "data/text_files/Java.txt": "Java is a versatile programming language. It is used for building enterprise applications, mobile apps, and large-scale systems.",
    "data/text_files/JavaScript.txt": "JavaScript is a popular programming language for web development. It is used to create interactive and dynamic web pages."
}

for file_path, text in sample_text.items():
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(text)   
```

### TextLoader-Read Single File

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("data/text_files/Python.txt",encoding="utf-8")
documents = loader.load()
print(documents)
```

### DirectoryLoader- Multiple Text Files

```python
from langchain_community.document_loaders import DirectoryLoader

directory_loader = DirectoryLoader("data/text_files", 
                                   glob="**/*.txt",
                                     show_progress=True,
                                     loader_kwargs={"encoding":"utf-8"}
                                     )
documents = directory_loader.load()
print(documents)    
```

```python
print(f"Character text splitter")

char_splitter =CharacterTextSplitter(
    separator="\n",
    chunk_size=200, 
    chunk_overlap=20,
    length_function=len,
    
)

character_chunks = char_splitter.split_text(documents[2].page_content)
```

```python
documents
```

```python
print(f"RecursiveCharacterTextSplitter")

recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=20,
    length_function=len,
    separators=["\n\n", "\n", " ", ""]
)
recursive_chunks = recursive_splitter.split_text(documents[2].page_content)
print(f"Created {len(recursive_chunks)} chunks using RecursiveCharacterTextSplitter")
print(f"First chunk: {recursive_chunks[0]}")
```

```python
print(f"TokenTextSplitter")
token_splitter = TokenTextSplitter(
    chunk_size=200,
    chunk_overlap=20,
    
        
)
token_chunks = token_splitter.split_text(documents[2].page_content) 
print(f"Created {len(token_chunks)} chunks using TokenTextSplitter")
print(f"First chunk: {token_chunks[0]}")
```

```python
chunks = token_splitter.split_documents(documents)
print(f"Created {len(chunks)} chunks using TokenTextSplitter on documents")

chunks = recursive_splitter.split_documents(documents)
print(f"Created {len(chunks)} chunks using RecursiveCharacterTextSplitter on documents")

chunks = char_splitter.split_documents(documents)
print(f"Created {len(chunks)} chunks using CharacterTextSplitter on documents")
```
