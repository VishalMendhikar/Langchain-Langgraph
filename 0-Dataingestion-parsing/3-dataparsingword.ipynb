```python
from langchain_community.document_loaders import (
    Docx2txtLoader,
    UnstructuredWordDocumentLoader)

```

```python
print(f"Docx2txtLoader")

try:
    docx_loader = Docx2txtLoader("data\Word_files\Machine_learning.docx")
    docx_docs = docx_loader.load()
    print(f"Number of documents loaded with Docx2txtLoader: {len(docx_docs)}")
    print(f"First document content: {docx_docs[0].page_content[:200]}")  # Print the first 200 characters of the first document
    print(f"metadata: {docx_docs[0].metadata}") # Print metadata of the first document
except Exception as e:
    print(f"Error loading Word document with Docx2txtLoader: {e}")
```

```python
docx_docs
```

```python
print(f"UnstructuredWordDocumentLoader")

try:
    unstructured_loader = UnstructuredWordDocumentLoader("data\Word_files\Machine_learning.docx",mode = "elements")
    unstructured_docs = unstructured_loader.load()
    print(f"Number of documents loaded with UnstructuredWordDocumentLoader: {len(unstructured_docs)}")
    print(f"First document content: {unstructured_docs[0].page_content[:200]}")  # Print the first 200 characters of the first document
    print(f"metadata: {unstructured_docs[0].metadata}") # Print metadata of the first document  
except Exception as e:
    print(f"Error loading Word document with UnstructuredWordDocumentLoader: {e}")
```
