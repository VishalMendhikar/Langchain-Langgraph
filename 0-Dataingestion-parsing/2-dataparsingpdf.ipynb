```python
from langchain_core.documents import Document
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from typing import List

class customPDFProcessor:
    def __init__(self, chunk_size=1000, chunk_overlap=100):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=self.chunk_size, 
            chunk_overlap=self.chunk_overlap,
            separators=[" ", "\n", "\n\n"] # Standard separators for better splitting
        )

    def process_pdf(self, file_path: str) -> List[Document]:
        loader = PyPDFLoader(file_path)
        pages = loader.load() # Returns a list of Document objects

        # Using split_documents is cleaner and preserves existing metadata
        chunks = self.text_splitter.split_documents(pages)
        
        # If you need to add custom metadata fields like 'chunk_method' 
        # specifically, you can update them after splitting:
        for chunk in chunks:
            chunk.metadata.update({
                "chunk_method": "RecursiveCharacterTextSplitter",
                "char_count": len(chunk.page_content)
            })
            
        return chunks
```

```python
preprocessor = customPDFProcessor()
preprocessor
```

```python
try:
    processed_docs = preprocessor.process_pdf("data/pdf_files/attention.pdf")
    print(f"Number of processed chunks: {len(processed_docs)}")
    print(f"First chunk content: {processed_docs[0].page_content[:50]}")  # Print the first 50 characters of the first chunk
    print(f"First chunk metadata: {processed_docs[0].metadata}")
except Exception as e:
    print(f"Error processing PDF: {e}")
```
